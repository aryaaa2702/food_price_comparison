```python
import requests
import json

url = "https://dummyjson.com/recipes"
response = requests.get(url)

print("Status Code:", response.status_code)
data = response.json()

print(json.dumps(data["recipes"][:2], indent=4))
```

```python
import pandas as pd
df= pd.DataFrame(data["recipes"])
df.head()
```

```python
df.to_csv("../data/raw/restaurant_data_raw.csv", index=False)
```

```python
df.columns
```

```python
list_columns = ["ingredients", "instructions", "cuisine"]  # adjust based on df.columns

for col in list_columns:
    if col in df.columns:
        df[col] = df[col].astype(str)
df=df.drop_duplicates()
df=df.reset_index(drop=True)
df=df.fillna({"rating":df["rating"].mean()})
```

```python
list_columns = []

for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, list)).any():
        list_columns.append(col)

list_columns
```

```python
# Convert list-type columns to strings
for col in ['tags', 'mealType']:
    df[col] = df[col].astype(str)

# Now drop duplicates
df = df.drop_duplicates()

# Reset index
df = df.reset_index(drop=True)

df.head()
```

```python
# Fill missing numerical columns with mean
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

# Fill missing categorical columns with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df["difficulty"] = df["prepTimeMinutes"].apply(
    lambda x: "Easy" if x < 20 else "Medium" if x < 45 else "Hard"
)
```

```python
df["calorie_density"] = df["caloriesPerServing"] / df["servings"]
```

```python
df["is_healthy"] = df["caloriesPerServing"].apply(lambda x: "Yes" if x < 400 else "No")
```

```python
df.to_csv("../data/processed/restaurant_data_clean.csv", index=False)
```
